```python
iris_df = sns.load_dataset('iris')
print("First 5 rows of the Iris DataFrame:")
print(iris_df.head())
```

```python
features = iris_df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
target = iris_df['species']

print("Features (first 5 rows):")
print(features.head())
print("\nTarget (first 5 rows):")
print(target.head())
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

print("Scaled Features (first 5 rows):")
print(scaled_features[:5])
```

```python
pca = PCA(n_components=2)
x_pca = pca.fit_transform(scaled_features)

print("Shape of PCA-transformed features:", x_pca.shape)
print("PCA-transformed features (first 5 rows):\n", x_pca[:5])
```

```python
pca_df = pd.DataFrame(data = x_pca, columns = ['Principal Component 1', 'Principal Component 2'])
pca_df['species'] = target

print("PCA DataFrame (first 5 rows):")
print(pca_df.head())
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x='Principal Component 1',
    y='Principal Component 2',
    hue='species',
    data=pca_df,
    s=80
)
plt.title('2 Component PCA of Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Species')
plt.grid(True)
plt.show()
```
